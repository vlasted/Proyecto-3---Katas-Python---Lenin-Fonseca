## Ejercicio 1

Escribe una función que reciba una cadena de texto como parámetro y devuelva un diccionario con las frecuencias de cada letra en la cadena. Los espacios no deben ser considerados.

### Nota explicativa
Se crea un diccionario para contar cuántas veces aparece cada letra. Primero se eliminan los espacios y se pasa el texto a minúsculas para que el conteo sea más uniforme. Después se recorre cada letra y se suma una unidad si ya existe en el diccionario.

```python
def frecuencia_letras(texto):
    frecuencias = {}
    texto_limpio = texto.replace(" ", "").lower()

    for letra in texto_limpio:
        if letra in frecuencias:
            frecuencias[letra] += 1
        else:
            frecuencias[letra] = 1

    return frecuencias
```

```python
print("#1", frecuencia_letras("Hola Antonio"))
```

## Ejercicio 3

Escribe una función que tome una lista de palabras y una palabra objetivo como parámetros. La función debe devolver una lista con todas las palabras de la lista original que contengan la palabra objetivo.

### Nota explicativa
Se busca si una palabra objetivo está contenida dentro de cada palabra de la lista. Para evitar problemas con mayúsculas y minúsculas, se convierte todo a minúsculas antes de comparar.

```python
def buscar_palabras(lista_palabras, objetivo):
    objetivo = objetivo.lower()
    return [palabra for palabra in lista_palabras if objetivo in palabra.lower()]
```

## Ejercicio 4

Genera una función que calcule la diferencia entre los valores de dos listas. Usa la función `map()`.

### Nota explicativa
Se calcula la diferencia entre dos listas posición por posición usando `map()`. Antes de hacerlo, se comprueba que ambas listas tengan la misma longitud para evitar errores.

```python
def diferencia_listas(lista1, lista2):
    if len(lista1) != len(lista2):
        raise ValueError("Las dos listas deben tener la misma longitud.")
    return list(map(lambda x, y: x - y, lista1, lista2))
```

```python
print("#4", diferencia_listas([10, 20, 30], [1, 2, 3]))
```

## Ejercicio 5

Escribe una función que tome una lista de números como parámetro y un valor opcional `nota_aprobado` (por defecto 5). La función debe calcular la media y devolver una tupla con la media y el estado.

### Nota explicativa
Se calcula la media de una lista de números y después se compara con la nota mínima para aprobar. La función devuelve una tupla con dos valores: la media y el estado final.

```python
def calcular_media_y_estado(lista_numeros, nota_aprobado=5):
    if len(lista_numeros) == 0:
        return (0, "suspenso")

    media = sum(lista_numeros) / len(lista_numeros)
    estado = "aprobado" if media >= nota_aprobado else "suspenso"
    return (media, estado)
```

```python
print("#5", calcular_media_y_estado([6, 7, 8]))
```

## Ejercicio 6

Escribe una función que calcule el factorial de un número de manera recursiva.

### Nota explicativa
Se usa recursividad. La función se llama a sí misma hasta llegar al caso base, que ocurre cuando el número es 0 o 1. También se controla que no se introduzcan números negativos.

```python
def factorial_recursivo(numero):
    if numero < 0:
        raise ValueError("El factorial no está definido para números negativos.")
    if numero == 0 or numero == 1:
        return 1
    return numero * factorial_recursivo(numero - 1)
```

```python
print("#6", factorial_recursivo(5))
```

## Ejercicio 7

Genera una función que convierta una lista de tuplas a una lista de strings. Usa la función `map()`.

### Nota explicativa
Se usa `map()` para transformar cada tupla en un string. Para ello, cada elemento de la tupla se convierte a texto y luego se une con `" ".join(...)`.

```python
def tuplas_a_strings(lista_tuplas):
    return list(map(lambda tupla: " ".join(map(str, tupla)), lista_tuplas))
```

Escribe un programa que pida al usuario dos números e intente dividirlos. Si el usuario ingresa un valor no numérico o intenta dividir por cero, maneja esas excepciones de manera adecuada y muestra un mensaje indicando si la división fue exitosa o no.

### Nota explicativa

## Ejercicio 9

Escribe una función que tome una lista de nombres de mascotas como parámetro y devuelva una nueva lista excluyendo ciertas mascotas prohibidas en España. La lista de mascotas a excluir es:
`["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]`

### Nota explicativa
Se usa `filter()` para eliminar de la lista las mascotas prohibidas. La función devuelve una nueva lista solo con las mascotas permitidas.

```python
def filtrar_mascotas_permitidas(lista_mascotas):
    prohibidas = ["Mapache", "Tigre", "Serpiente Pitón", "Cocodrilo", "Oso"]
    return list(filter(lambda mascota: mascota not in prohibidas, lista_mascotas))
```

```python
print("#9", filtrar_mascotas_permitidas(["Perro", "Mapache", "Gato", "Oso", "Loro"]))
```

## Ejercicio 10

Escribe una función que reciba una lista de números y calcule su promedio. Si la lista está vacía, lanza una excepción personalizada y maneja el error adecuadamente.

### Nota explicativa
Se calcula el promedio de una lista, pero si la lista está vacía se lanza una excepción personalizada llamada `ListaVaciaError`. De esta forma se controla mejor ese caso especial.

```python
class ListaVaciaError(Exception):
    pass

def calcular_promedio(lista_numeros):
    if len(lista_numeros) == 0:
        raise ListaVaciaError("La lista está vacía.")

    return sum(lista_numeros) / len(lista_numeros)
```

```python
try:
    print("#10", calcular_promedio([5, 10, 15]))
except ListaVaciaError as error:
    print("#10", error)
```

## Ejercicio 12

Genera una función que, al recibir una frase, devuelva una lista con la longitud de cada palabra. Usa la función `map()`.

### Nota explicativa
Se separa la frase en palabras usando `split()`. Después se usa `map()` para obtener la longitud de cada palabra y devolverlo como una lista.

```python
def longitud_palabras(frase):
    palabras = frase.split()
    return list(map(len, palabras))
```

## Ejercicio 13

Genera una función que, para un conjunto de caracteres, devuelva una lista de tuplas con cada letra en mayúsculas y minúsculas. Las letras no pueden estar repetidas. Usa la función `map()`.

### Nota explicativa
Primero se eliminan los caracteres repetidos usando `set()`. Después se transforma cada carácter en una tupla formada por su versión en mayúscula y en minúscula mediante `map()`.

```python
def letras_mayus_minus(caracteres):
    letras_unicas = list(set(caracteres))
    return list(map(lambda letra: (letra.upper(), letra.lower()), letras_unicas))
```

```python
print("#13", letras_mayus_minus("abracadabra"))
```

## Ejercicio 14

Crea una función que retorne las palabras de una lista que comiencen con una letra en específico. Usa la función `filter()`.

### Nota explicativa
Se usa `filter()` para devolver únicamente las palabras que empiecen por una letra concreta. Antes de comparar, tanto la letra como las palabras se convierten a minúsculas.

```python
def palabras_por_letra(lista_palabras, letra):
    letra = letra.lower()
    return list(filter(lambda palabra: palabra.lower().startswith(letra), lista_palabras))
```

```python
print("#14", palabras_por_letra(["casa", "coche", "perro", "camino"], "c"))
```

## Ejercicio 16

Escribe una función que tome una cadena de texto y un número entero `n` como parámetros y devuelva una lista de todas las palabras que sean más largas que `n`. Usa la función `filter()`.

### Nota explicativa
Primero se separa el texto en palabras con `split()`. Después se utiliza `filter()` para quedarse solo con las palabras cuya longitud sea mayor que el valor `n`.

```python
def palabras_mas_largas_que(texto, n):
    palabras = texto.split()
    return list(filter(lambda palabra: len(palabra) > n, palabras))
```

## Ejercicio 17

Crea una función que tome una lista de dígitos y devuelva el número correspondiente. Por ejemplo, `[5,7,2]` corresponde al número `572`. Usa la función `reduce()`.

### Nota explicativa
Se usa `reduce()` para ir construyendo el número completo a partir de los dígitos. La lógica es multiplicar el acumulado por 10 y sumarle el siguiente dígito, de forma que el número se vaya formando paso a paso.

```python
def digitos_a_numero(lista_digitos):
    return reduce(lambda acumulado, digito: acumulado * 10 + digito, lista_digitos, 0)
```

```python
print("#17", digitos_a_numero([5, 7, 2]))
```

## Ejercicio 20

Para una lista con elementos de tipo integer y string, obtén una nueva lista solo con los valores int. Usa la función `filter()`.

### Nota explicativa
Se utiliza `filter()` junto con `isinstance()` para comprobar el tipo de cada elemento de la lista. Así se devuelven únicamente los valores que sean enteros.

```python
def solo_enteros(lista_mixta):
    return list(filter(lambda elemento: isinstance(elemento, int), lista_mixta))
```
